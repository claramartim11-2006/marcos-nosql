```python
df = pd.DataFrame(list(colecao_atendimentos.find()))
```

```python
nulos = df.isnull().sum()
```

```python
faturamento_total = float(df['valor'].sum())
```

```python
faturamento_por_cidade = {k: float(v) for k, v in df.groupby('cidade')['valor'].sum().items()}
```

```python
faturamento_por_servico = {k: float(v) for k, v in df.groupby('servico')['valor'].sum().items()}
```

```python
ticket_medio_geral = float(df['valor'].mean())
```

```python
ticket_medio_por_cidade = {k: float(v) for k, v in df.groupby('cidade')['valor'].mean().round(2).items()}
```

```python
servico_mais_lucrativo = str(df.groupby('servico')['valor'].sum().idxmax())
```

```python
cidade_maior_faturamento = str(df.groupby('cidade')['valor'].sum().idxmax())
```
